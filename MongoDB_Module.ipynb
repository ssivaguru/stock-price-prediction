```python
import datetime                            # Imports datetime library

import pymongo
from pymongo import MongoClient
from pymongo.errors import BulkWriteError

myclient = pymongo.MongoClient("mongodb://localhost:27017/")

mydb = myclient["stock_db"]
mycol = mydb["stock"]
mycol.drop()
print(mydb.list_collection_names())
#insert multiple documents
mylist = [
  { "stock_name": "Samsung","model_Path": "ModelPath_1", "prediction": "1.5",  "requesting_users": "", "state": "Trained"},
  { "stock_name": "Jio", "model_Path": "ModelPath_2", "prediction": "9.5", "requesting_users": "", "state": "Trained"}
]

myDB = mycol.insert_many(mylist)

print("\nFind entries with stock_name")
for dbIdx in mycol.find({},{ "stock_name": "Samsung"}):
  print(dbIdx)

print("\nDelete entries with stock_name")
myquery = { "stock_name": "Samsung" }
#Delete one document
mycol.delete_one(myquery)

#Add new entry
mylist = [
  { "stock_name": "Samsung","model_Path": "ModelPath_1", "prediction": "1.5",  "requesting_users": "", "state": "In-progress"},
  { "stock_name": "Airtel", "model_Path": "ModelPath_2", "prediction": "9.5", "requesting_users": "", "state": "In-progress"}
]
myDB = mycol.insert_many(mylist)
#print(myDB.inserted_ids)

#Create coolection for User auth
mycol1 = mydb["user"]
mycol1.drop()
print("\n",mydb.list_collection_names())
#insert multiple documents
mylist1 = [
  { "user_name": "ABC", "email": "abc@gmail.com"},
  { "user_name": "MNOP", "email": "mnop@gmail.com"}
]

myDB = mycol1.insert_many(mylist1)
print("\n",myDB.inserted_ids)
print("\n",mydb.list_collection_names())
print("\n51:find entries with user_name")
for dbIdx in mycol1.find({},{ "user_name": "MNOP"}):
  print("\n",dbIdx)

#Update an entry
print("\nUpdate entry for stock_name")
myquery = { "stock_name": "Samsung" }
newvalues = { "$set": { "state": "Trained" } }
mycol.update_one(myquery, newvalues)

#Find an entry
print("\nFind entry for stock_name")
myquery = { "stock_name": "Samsung" }
mydoc = mycol.find(myquery)
for dbIdx in mydoc:
  print(dbIdx)

print("\nFind sorted entry for stock_name")
mydoc = mycol.find().sort("name", -1)
for dbIdx in mydoc:
  print("\n",dbIdx)
mydoc = mycol1.find().sort("name", -1)
for dbIdx in mydoc:
  print("\n",dbIdx)
```
